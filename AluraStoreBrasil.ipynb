### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. An√°lise do faturamento



```python
# Somando o faturamento de cada loja separadamente
# O :,.2f serve para colocar v√≠rgula e 2 casas decimais (estilo dinheiro)

print("--- üí∞ Faturamento Total: LOJA 1 ---")
fat1 = loja['Pre√ßo'].sum()
print(f"R$ {fat1:,.2f}")

print("\n--- üí∞ Faturamento Total: LOJA 2 ---")
fat2 = loja2['Pre√ßo'].sum()
print(f"R$ {fat2:,.2f}")

print("\n--- üí∞ Faturamento Total: LOJA 3 ---")
fat3 = loja3['Pre√ßo'].sum()
print(f"R$ {fat3:,.2f}")

print("\n--- üí∞ Faturamento Total: LOJA 4 ---")
fat4 = loja4['Pre√ßo'].sum()
print(f"R$ {fat4:,.2f}")
```

# 2. Vendas por Categoria

```python
# Contando as categorias para cada loja
# O comando .value_counts() conta quantos produtos de cada tipo foram vendidos

print("--- üì¶ Categorias: LOJA 1 ---")
display(loja['Categoria do Produto'].value_counts())

print("\n--- üì¶ Categorias: LOJA 2 ---")
display(loja2['Categoria do Produto'].value_counts())

print("\n--- üì¶ Categorias: LOJA 3 ---")
display(loja3['Categoria do Produto'].value_counts())

print("\n--- üì¶ Categorias: LOJA 4 ---")
display(loja4['Categoria do Produto'].value_counts())
```

# 3. M√©dia de Avalia√ß√£o das Lojas


```python
# Calculando a m√©dia das notas (estrelas)

print("--- ‚≠ê Nota M√©dia: LOJA 1 ---")
nota1 = loja['Avalia√ß√£o da compra'].mean()
print(f"{nota1:.2f} / 10")

print("\n--- ‚≠ê Nota M√©dia: LOJA 2 ---")
nota2 = loja2['Avalia√ß√£o da compra'].mean()
print(f"{nota2:.2f} / 10")

print("\n--- ‚≠ê Nota M√©dia: LOJA 3 ---")
nota3 = loja3['Avalia√ß√£o da compra'].mean()
print(f"{nota3:.2f} / 10")

print("\n--- ‚≠ê Nota M√©dia: LOJA 4 ---")
nota4 = loja4['Avalia√ß√£o da compra'].mean()
print(f"{nota4:.2f} / 10")
```

# 4. Produtos Mais e Menos Vendidos

```python
# --- LOJA 1 ---
print("--- üèÜ Loja 1: Mais Vendidos (Top 5) ---")
display(loja['Produto'].value_counts().head(5))
print("--- üìâ Loja 1: Menos Vendidos (Bottom 5) ---")
display(loja['Produto'].value_counts().tail(5))

# --- LOJA 2 ---
print("\n--- üèÜ Loja 2: Mais Vendidos (Top 5) ---")
display(loja2['Produto'].value_counts().head(5))
print("--- üìâ Loja 2: Menos Vendidos (Bottom 5) ---")
display(loja2['Produto'].value_counts().tail(5))

# --- LOJA 3 ---
print("\n--- üèÜ Loja 3: Mais Vendidos (Top 5) ---")
display(loja3['Produto'].value_counts().head(5))
print("--- üìâ Loja 3: Menos Vendidos (Bottom 5) ---")
display(loja3['Produto'].value_counts().tail(5))

# --- LOJA 4 ---
print("\n--- üèÜ Loja 4: Mais Vendidos (Top 5) ---")
display(loja4['Produto'].value_counts().head(5))
print("--- üìâ Loja 4: Menos Vendidos (Bottom 5) ---")
display(loja4['Produto'].value_counts().tail(5))
```

# 5. Frete M√©dio por Loja

```python
# Calculando a m√©dia do frete

print("--- üöö Frete M√©dio: LOJA 1 ---")
frete1 = loja['Frete'].mean()
print(f"R$ {frete1:.2f}")

print("\n--- üöö Frete M√©dio: LOJA 2 ---")
frete2 = loja2['Frete'].mean()
print(f"R$ {frete2:.2f}")

print("\n--- üöö Frete M√©dio: LOJA 3 ---")
frete3 = loja3['Frete'].mean()
print(f"R$ {frete3:.2f}")

print("\n--- üöö Frete M√©dio: LOJA 4 ---")
frete4 = loja4['Frete'].mean()
print(f"R$ {frete4:.2f}")
```

#Gr√°ficos

```python
import matplotlib.pyplot as plt

# 1. Organizando os dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [fat1, fat2, fat3, fat4]
fretes_medios = [frete1, frete2, frete3, frete4]
avaliacoes = [nota1, nota2, nota3, nota4]

plt.figure(figsize=(16, 10))

# --- GR√ÅFICO 1: Faturamento ---
plt.subplot(2, 2, 1)
plt.bar(lojas, faturamentos, color='#4e79a7')
plt.title('Faturamento Total')
plt.ylabel('Reais (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# --- GR√ÅFICO 2: Frete ---
plt.subplot(2, 2, 2)
plt.barh(lojas, fretes_medios, color='#e15759')
plt.title('Custo Medio do Frete')
plt.xlabel('Reais (R$)')

# --- GR√ÅFICO 3: Avalia√ß√£o ---
plt.subplot(2, 2, 3)
plt.bar(lojas, avaliacoes, color='#f28e2b')
plt.title('Avaliacao Media dos Clientes (0 a 10)')
plt.ylim(0, 10)

for i, nota in enumerate(avaliacoes):
    plt.text(i, nota + 0.1, f'{nota:.2f}', ha='center', fontweight='bold')

# --- GR√ÅFICO 4: Categorias ---
plt.subplot(2, 2, 4)
fatias = loja['Categoria do Produto'].value_counts().head(5)
plt.pie(fatias, labels=fatias.index, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink', 'silver'])
plt.title('Top 5 Categorias (Loja 1)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 7))

# --- MAPA DA LOJA 1 (A do Frete Caro) ---
plt.subplot(1, 2, 1)
# c='Frete' define que a cor muda conforme o pre√ßo
# cmap='Reds' usa tons de vermelho (quanto mais escuro, mais caro)
grafico1 = plt.scatter(loja['lon'], loja['lat'], c=loja['Frete'], cmap='Reds', alpha=0.6)
plt.colorbar(grafico1, label='Valor do Frete (R$)')
plt.title(' Mapa de Frete: LOJA 1 (Frete Caro)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# --- MAPA DA LOJA 4 (A do Frete Barato) ---
plt.subplot(1, 2, 2)
# Vamos usar a mesma escala de cor para ser justo na compara√ß√£o
grafico4 = plt.scatter(loja4['lon'], loja4['lat'], c=loja4['Frete'], cmap='Reds', alpha=0.6)
plt.colorbar(grafico4, label='Valor do Frete (R$)')
plt.title(' Mapa de Frete: LOJA 4 (Frete Barato)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```
